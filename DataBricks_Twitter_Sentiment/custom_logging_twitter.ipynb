Program for creating log file.Then, using it for logging notebook execution statements.
* configuration level -> DEBUG.
* file path -> '/tmp/'
* Use commented code to get separate log file with different timestamp.

```
import logging
import datetime
import time


file_date = datetime.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d-%H-%M-%S')
p_filename = "custom_log"+file_date+".log"

p_dir = '/tmp/'
#p_filename = 'custom_log.log'
p_logfile = p_dir + p_filename
print(p_logfile)


logger = logging.getLogger('custom_log')
logger.setLevel(logging.DEBUG)

f_handler = logging.FileHandler(p_logfile, mode = 'a')


c_handler = logging.StreamHandler()
c_handler.setLevel(logging.DEBUG)


f_format = logging.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s")

f_handler.setFormatter(f_format)
c_handler.setFormatter(f_format)

if (logger.hasHandlers()):
  logger.handlers.clear()
  
logger.addHandler(f_handler)
logger.addHandler(c_handler)
```
